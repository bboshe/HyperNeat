```python
%matplotlib inline
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
location = "./saves2/"

history = []
for file in os.listdir(location):
    if file.startswith("lc"):
        history.append(pd.read_csv(os.path.join(location, file)))
```

```python
for h in history:
    plt.plot(h['generation'], h['max_fitness'])
```
